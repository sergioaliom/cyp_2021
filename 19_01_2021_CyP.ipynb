#Slicing (Continuacion)
```
[ inicio :  stop : incremento ]
```
Notas:
1. EL stop no es incluido.
2. El incremento se puede modificar
3. Soporta el indexado alternativo (Negativos)


````
frase = "Como hacer a la gente mas positiva?, muy facil, has que pierdan un electron!"
print(frase)
#imprimir "ace" de la palabra hacer 
print(frase[6:9:1])


#Valores por defecto
print(frase[:5:])

#Incremento
print(frase[: :2]) #Toda la frase
print(frase[-1: :-1]) #Inverso

nombre = "Juan Jose"
print(nombre[1::-1])
primer = nombre[:4:]
print(primer)
print(nombre[-9::1])
````

# Metodos de la clase string (str)

En python todo es un objeto (P.O.O.), eso significa que cada tipo de dato trae consigo metodos (funciones) incluida.

Para identificar los metodos disponibles se puede usar una de dos funciones 
1. dir(str) #Resumen de los metodos de esa clase.
2. help(str) #Informacion completa de los metodos de esa clase.

````
nombre = "Juan Jose"
print(nombre)
print(dir(str))

metodos = dir(str)
for metodos in metodos:
  print(metodos)
````

Los nombres de metodos que inician con doble guion bajo, significa que son metodos privados (POO), y ni tendrias porque usarlos como programador consumidor.

Los demas metodos son publicos y son de libre uso.

````
print(nombre)
print( nombre.islower())
nombre = "PEDRO FERNANDEZ"
print(nombre.capitalize())
print(nombre.endswith("DEZ"))
print(nombre.replace("E","*"))
ind_e = nombre.find("E")
print(ind_e)
nombre = nombre[:ind_e:] + "@" + nombre[ind_e+1::]
print(nombre)
print(nombre.replace("E","@",0))
print(nombre.lower())
nombre = "jose"
print(nombre.upper())
````

````
# Metodos para pricesar Strings

numeros = "  2,  1, 2,     3,       6, 53, 45,  5                "
print(numeros)
#Como extraigo los numeros de la cadena
#1. Eliminar los espacios al principio y al final con strip()
numeros = numeros.strip()
print(numeros)
#2, Separo por comas
numeros = numeros.split(",")
print(numeros)
#3, Con un for e indices limpiar cadena del la lista
for indice in range(len(numeros)):
  numeros[indice] = numeros[indice].strip()
print(numeros)
#4, Convertirlo a tipo entero 
for indice in range( len(numeros)):
  numeros[indice]= int( numeros[indice])
print(numeros)
````
